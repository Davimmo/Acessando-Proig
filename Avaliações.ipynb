```python
import time
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
driver=webdriver.Chrome()
from selenium.webdriver.common.alert import Alert
alert=Alert(driver)

driver.get('https://www.educacao.petrolina.pe.gov.br/')
```

```python
def gravar_mudanças():
carlos=driver.find_elements(By.TAG_NAME,'span')
for elementos in carlos:
    if str(elementos.text)=='Gravar':
        elementos.click()
        break
```

```python
bercariomulti=[
'ALANA BEATRIZ DA SILVA LIMA',
'CRYSTAL CECÍLIA ARAUJO DA SILVA',
'ELISA ANGELO GOMES',
'ESTHER OLIVEIRA NOVAIS',
'JAILSON HEITOR SANTANA DA SILVA',
'LORENZO DE SOUZA FERREIRA',
'MARIA YASMIN TOMÁS DOS SANTOS',
'NOÁN VITOR BATISTA SILVA',
'RICHELY DOS SANTOS ALMEIDA',
'RUAN LORENZO DA SILVA',
'YAGO DE SOUZA DO E'
]


testando=[]


count=0

for alunos in bercariomulti:


    buscar_nomes=driver.find_elements(By.TAG_NAME,'td')
    
    for buscado in buscar_nomes:
        testando.append(str(buscado.text))
    
    if str(alunos) in testando:
        for numeros in buscar_nomes:
            if str(numeros.text)==str(alunos):

                numeros.click()
                time.sleep(2)
    




            davi=driver.find_elements(By.TAG_NAME,'label')
            for coisas in davi:
                if str(coisas.text)=='DESENVOLVIDO':
                    coisas.click()

                    





            carlos=driver.find_elements(By.TAG_NAME,'span')
            for botão in carlos:
                if str(botão.text)=='Gravar':
                    botão.click()
                    time.sleep(2)
                    break

    
            count+=1  

    
    testando=[]
    time.sleep(4)


```
